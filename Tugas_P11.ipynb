```python
import pandas as pd

df_excel = pd.read_excel('disperkim-od_16985_jumlah_produksi_sampah_berdasarkan_kabupatenkota_v3_data.xlsx', sheet_name='data')
df_excel
```

```python
df_filter = df_excel[['nama_kabupaten_kota','jumlah_produksi_sampah','satuan','tahun']]
df_filter
```

```python
sampah_2021 = []

for index, row in df_excel.iterrows():
    if row['tahun'] == 2021:
        sampah_2021.append(row['jumlah_produksi_sampah'])

total_sampah_2021 = sum(sampah_2021)

dfsampah2021 = pd.DataFrame({'tahun': [2021], 'jumlah produksi sampah': [total_sampah_2021]})

dfsampah2021
```

```python
print("List Jumlah Produksi Sampah Dari Kota Per Tahun")

# Inisialisasi dictionary untuk menampung total sampah per tahun
jumlah_sampah_per_tahun = {
    2015: 0,
    2016: 0,
    2017: 0,
    2018: 0,
    2019: 0,
    2020: 0,
    2021: 0,
    2022: 0,
    2023: 0
}

# Iterasi melalui DataFrame untuk mengisi total sampah per tahun
for index, row in df_excel.iterrows():
    tahun = row['tahun']
    if tahun in jumlah_sampah_per_tahun:
        jumlah_sampah_per_tahun[tahun] += row['jumlah_produksi_sampah']

# Membuat DataFrame untuk menampilkan hasil
df_total_sampah = pd.DataFrame(list(jumlah_sampah_per_tahun.items()), columns=['Tahun', 'Total Produksi Sampah'])

df_total_sampah
```

```python
total_wilayah_tahun = {}
for index, row in df_excel.iterrows():
    wilayah = row['nama_kabupaten_kota']
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    
    if wilayah not in total_wilayah_tahun:
        total_wilayah_tahun[wilayah] = {}
    
    if tahun in total_wilayah_tahun[wilayah]:
        total_wilayah_tahun[wilayah][tahun] += jumlah
    else:
        total_wilayah_tahun[wilayah][tahun] = jumlah

result = []
for wilayah, tahun_data in total_wilayah_tahun.items():
     for tahun, jumlah in tahun_data.items():
         result.append([wilayah, tahun, jumlah])
         
df_result = pd.DataFrame(result, columns=('Wilayah','Tahun','Jumlah Sampah (Ton)'))
df_result
         
```
